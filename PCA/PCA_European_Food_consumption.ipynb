```python
## Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
class pca_q1:
    
    def __init__(self, k):
        self.k = k
    
    def pca_q1_country(self,data):
        k = self.k
        
        #Finding the mean of the data and minusing it from the data set.
        # Axis 0 is row mean
        # Axis 1 is column mean
        data_mean = np.mean(data, axis = 0).to_numpy()
        c = data.T - data_mean.reshape(-1,1)
        
        #SVD 
        U,E,V = np.linalg.svd(c)
        
        #z^i
        dim_list = []
        
        #Reduction of representation of data points
        for i in range(k):
            dim = (U[:,i].T@c)/(E[i]**.5)
            dim_list.append(dim)
        
        return dim_list
    
    def plot_2d(self,pca_data,data,title):
        
        #Labels
        labels = data.T.columns.to_numpy()
        
        # Setting up plot with data)
        fig = plt.figure(figsize=(10,10))
        ax = plt.scatter(pca_data[0],pca_data[1])
        
        for i in range(len(labels)):
            y = labels[i]
            plt.text(pca_data[0][i]-.4,pca_data[1][i]+.2,y)
        plt.title(title)
        
        return fig
```

```python
data = pd.read_csv('data/food-consumption.csv')
data.set_index('Country', inplace = True)
    
pca = pca_q1(2)
pca_data = pca.pca_q1_country(data)
fig = pca.plot_2d(pca_data,data,'PCA on country\'s food consumption in Europe')
```

```python
pca_data = pca.pca_q1_country(data.T)
fig = pca.plot_2d(pca_data,data.T,'PCA on foods consumed in Europe')
```
